```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np   # use numpy to create some toy data
x = np.linspace(0, 5, 11)
y = x ** 2  # y = x * x

print('x: ', type(x), x)
print('y: ', type(y), y)
```

```python
len(x), len(y)
```

```python
# basic matplotlib plot
plt.plot(y, x, 'r^--')
plt.xlabel('y')
plt.ylabel('x')
plt.title('Insert Title Here')
```

```python
plt.subplot(1, 2, 1)  # work on plot 1
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)  # work on plot 2
plt.plot(x, y, 'g-')
```

```python
plt.subplot(2, 1, 1)  # work on plot 1
plt.plot(x, y, 'r--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot 1')
plt.subplot(2, 1, 2)  # work on plot 2
plt.plot(x, y, 'g-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot 2')
```

```python
# Object oriented way to make plots

# Create a Figure object (empty canvas)
fig = plt.figure()

# Add set of axes to the figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot on that set of axes
axes.plot(x, y, 'b')
```

```python
# Create a Figure object (empty canvas)
fig = plt.figure()

# Add set of axes to the figure
axes = fig.add_axes([0.2, 0.2, 0.5, 0.5])

# Plot on that set of axes
axes.plot(x, y, 'b')
```

```python
x, y
```

```python
# create blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8, 0.8])  # main axes
axes2 = fig.add_axes([0.2,0.5, 0.4, 0.3])  # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_labels_axe1')
axes1.set_title('Axes 1 Title')

# Smaller Figure Axes 2 
axes2.plot(x, y, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_labels_axes')
axes2.set_title('Axes 2 Title')
```

```python
fig.savefig('example.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,0.5,0.5])
ax.plot(x, x**2, label='y=x^2')
ax.plot(x, x**3, label='y=x^3')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('function')
ax.legend(['power 2', 'power 3'],loc='upper right')
#ax.set_xlim([0,3])
```

# Seaborn

```python
import seaborn as sns
```

```python
sns.set_style('whitegrid')
```

```python
titantic = sns.load_dataset('titanic')
print(titantic.shape)
titantic.head()
```

```python
sns.jointplot(x='fare', y='age', data=titantic)
```

```python
sns.distplot(titantic['fare'])
```

```python
# plt.plot(x, y)
```

```python
plt.hist(titantic['fare'], bins=30)  # 
```

```python
sns.boxplot(x='class', y='age', data=titantic)
```

```python
sns.countplot(x='sex',data=titantic)
```

```python
sns.heatmap(titantic.corr(), cmap='coolwarm')
```
